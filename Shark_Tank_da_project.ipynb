```python
li=list(df.columns[-4:-11:-1])
print(li)
def sharks(name):
  for i in li:
    #if name==i[:-5]:
     #   continue
    shark=name+"_deal"
    deal=df[(df[i]==1) & (df[shark]==1)]
    print(name,len(deal)," deal with",i[:-5])
    print(f"amount per shark {name}",deal["amount_per_shark"])
    print(f"equity per shark {name}",deal["equity_per_shark"])  
```

```python
sharks("ashneer")
```

```python
sharks("namita")
```
